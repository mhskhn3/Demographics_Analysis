```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data.csv")
df
```

```python
top_cities = df.sort_values(by='Users', ascending=False).head(20)
top_cities
```

```python
top_cities = df.sort_values(by='Users', ascending=False).head(20)
top_cities = df.sort_values(by='Users', ascending=False).head(20)
plt.figure(figsize=(12, 8))
sns.barplot(y=top_cities['Town/City'], x=top_cities['Users'], hue=top_cities['Town/City'], palette='Spectral', legend=False)
plt.xlabel('Number of Users')
plt.ylabel('Town/City')
plt.title('Top 20 Cities by Number of Users')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(y=top_cities['Town/City'], x=top_cities['Engagement rate'],hue= top_cities['Town/City'], palette='Spectral')
plt.xlabel('Engagement Rate')
plt.ylabel('Town/City')
plt.title('Engagement Rate in Top 20 Cities by Number of Users')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(y=top_cities['Town/City'], x=top_cities['Engaged sessions per user'], hue= top_cities['Town/City'], palette='flare')
plt.xlabel('Engaged Sessions per User')
plt.ylabel('Town/City')
plt.title('Engaged Sessions per User in Top 20 Cities by Number of Users')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(y=top_cities['Town/City'], x=top_cities['Average engagement time'], hue= top_cities['Town/City'], palette='crest')
plt.xlabel('Average Engagement Time (seconds)')
plt.ylabel('Town/City')
plt.title('Average Engagement Time in Top 20 Cities by Number of Users')
plt.show()
```

```python
top_cities['Returning users'] = top_cities['Users'] - top_cities['New users']

top_cities_melted = top_cities.melt(id_vars='Town/City', value_vars=['New users', 'Returning users'],
                                    var_name='User Type', value_name='Count')

plt.figure(figsize=(14, 10))
sns.barplot(y='Town/City', x='Count', hue='User Type', data=top_cities_melted, palette='viridis')
plt.xlabel('Number of Users')
plt.ylabel('Town/City')
plt.title('New Users vs Returning Users in Top 20 Cities by Number of Users')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(y=top_cities['Town/City'], x=top_cities['Event count'],hue=top_cities['Town/City'] , palette='magma')
plt.xlabel('Event Count')
plt.ylabel('Town/City')
plt.title('Event Count in Top 20 Cities by Number of Users')
plt.show()
```

```python
# high-Engagement Users
high_engagement_users = df[(df['Engaged sessions per user'] > df['Engaged sessions per user'].mean()) &
                             (df['Average engagement time'] > df['Average engagement time'].mean())]

# new Users
new_users = df[df['New users'] > (0.5 * df['Users'])]

# returning Users
returning_users = df[df['Users'] - df['New users'] > (0.5 * df['Users'])]

# categorize cities into high, medium, and low user count segments based on quantiles
user_quantiles = df['Users'].quantile([0.33, 0.67])
low_user_cities = df[df['Users'] <= user_quantiles[0.33]]
medium_user_cities = df[(df['Users'] > user_quantiles[0.33]) & (df['Users'] <= user_quantiles[0.67])]
high_user_cities = df[df['Users'] > user_quantiles[0.67]]

# summary of segments
segments_summary = {
    "High Engagement Users": len(high_engagement_users),
    "New Users": len(new_users),
    "Returning Users": len(returning_users),
    "Low User Cities": len(low_user_cities),
    "Medium User Cities": len(medium_user_cities),
    "High User Cities": len(high_user_cities)
}

segments_summary
```
